```python
from schemas.screener import ScreenConfig,AgentScreener
config = ScreenConfig(policy_vector,3)
sc = AgentScreener(config)
screened_agents = sc.screen_agents()

from utils.general import generate_agent_details_string
details=generate_agent_details_string(screened_agents)
print(details)
```

```python
from events.generator import create_event_objects
eventlist = create_event_objects(policy_text="A new tax will be introduced starting from 25th of July for public transportation, this extra tax will be used to fund education of poor children",agent_details=details,llm=llm)
```

```python
print(f"Total Events: {len(eventlist)}\n")
for i, event in enumerate(eventlist, 1):
    print(f"Event {i}:")
    print(f"  ID: {event.event_id}")
    print(f"  Agent ID: {event.agent_id}")
    print(f"  Timeline: {event.timeline}")
    # event_description is a list, so we join it or print the first element
    desc = " ".join(event.event_description) if isinstance(event.event_description, list) else event.event_description
    print(f"  Description: {desc}")
    print("-" * 50)
```

```python
reactions = generate_base_reactions(events=eventlist, llm=llm)
```

```python
for i, r in enumerate(reactions, 1):
    print(f"--- Reaction {i} ---")
    print(f"Event ID : {r['event_id']}")
    print(f"Agent ID : {r['agent_id']}")
    print(f"Reaction : {r['base_reaction']}")
    print()
```

```python
from IPython.display import Image, display
from events.runner.base_reaction import _build_reaction_graph

# Build the reaction graph using the LLM instance
app = _build_reaction_graph(llm)

# Get the drawable graph from the compiled application
graph = app.get_graph()

try:
    # Attempt to generate a Mermaid diagram of the graph
    png_data = graph.draw_mermaid_png()
except Exception as e:
    print(f"Mermaid diagram generation failed: {e}")
    # Fallback to a standard PNG diagram if available
    png_data = graph.draw_png()

# Display the generated graph image
display(Image(data=png_data))
```
